```python
df_tweets = df.loc[:,['content','retweets','date','favorites']].sort_values('retweets',ascending=False)
df_tweets.head()
```

```python
df_tweets['year'] = [date[:4] for date in df_tweets['date']]
df_tweets.head()
```

```python
df['year'] = [date[:4] for date in df['date']]
df.head()
```

```python
sns.scatterplot(x=sorted(df.year), y=df.retweets)
```

```python
sns.scatterplot(x=sorted(df.year), y=df.favorites)
```
